3. Two real world scenarios are sentiment analysis and Q/A chatbot:-
*   In sentiment analysis, removing stopwords can change the meaning of a sentence because words like “not”, “no”, and “never” are important. For example, “This movie is not good” may become “movie good,” which gives the wrong (positive) sentiment.

*  In question answering or chatbots, stopwords help maintain the structure and intent of a sentence. For example, “What is the time now?” becomes “time now” after removal, making the query unclear and harder to understand.

Stemming is a simple process that just cuts off prefixes or suffixes, so it can sometimes produce incorrect or incomplete words, like “studies” becoming “studi.” On the other hand, lemmatization is more accurate because it understands the meaning of the word and converts it into a proper root form, such as “studies” becoming “study.

**Task 4: Token Analytics**

For each processed sentence, compute:
1. Total number of tokens
2. Number of unique tokens
3. Average token length


```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)

    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    return tokens


def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    return total_tokens, unique_tokens, round(avg_length, 2)

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


for i, sentence in enumerate(sentences, 1):
    tokens = preprocess_text(sentence)
    total, unique, avg_len = token_analytics(tokens)

    print(f"\n{i}. Sentence: {sentence}")
    print(f"   Tokens: {tokens}")
    print(f"   Total Tokens: {total}")
    print(f"   Unique Tokens: {unique}")
    print(f"   Avg Token Length: {avg_len}")
```

```python
import re

def preprocess_text(text):
    # Handle empty input
    if not text or text.strip() == "":
        return [], ""

    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)

    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Handle cases like only emojis / only numbers → becomes empty
    if len(tokens) == 0:
        return [], ""

    return tokens, ' '.join(tokens)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

test_cases = [
    "",
    "😍😍😂😂",
    "1234567890",
    "Hello 123 😊",
]

result = full_pipeline(test_cases)

print("Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```

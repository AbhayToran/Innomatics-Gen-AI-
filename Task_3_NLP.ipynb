## **Step 2**: Importing Libraries

We import the required classes for loading the model and handling text processing.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

## **Step 3**: Loading Pre-trained Model

We load the GPT-2 model from Hugging Face.

### **Why GPT-2?**
- General-purpose language model  
- Generates structured text  
- Works well with prompt-based control  



```python
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

print("GPT-2 Model loaded successfully!\n")
```

## **Understanding GPT-2**

GPT-2 is a transformer-based language model that generates text by predicting the next word in a sequence.

### **Key Components**:
- Tokenizer → Converts text into numerical tokens  
- Model → Predicts next sequence of words  



## **Step 4:** Chatbot Implementation

In this step, we implement the chatbot logic.

### **Features**:
- Accepts user input  
- Generates responses using GPT-2  
- Maintains conversation history  
- Uses prompt engineering  
- Supports continuous interaction  
- Includes exit condition  

```python
conversation_history = ""

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:

    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Conversation ended. Goodbye!")
        break

    # DYNAMIC PROMPT
    prompt = (
        "You are a knowledgeable and professional AI assistant.\n"
        "Give accurate, clear, and concise answers.\n"
        "Avoid repetition and irrelevant information.\n"
        "If unsure, say you are not certain.\n\n"
        f"Question: {user_input}\nAnswer:"
    )

    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate response
    output = model.generate(
        input_ids,
        max_length=input_ids.shape[1] + 80,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        temperature=0.2,
        top_k=40,
        top_p=0.9,
        repetition_penalty=1.7
    )

    response = tokenizer.decode(
        output[:, input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    # Clean response
    response = response.strip()
    response = response.split("Question:")[0]

    # Remove repeated lines
    sentences = list(dict.fromkeys(response.split(".")))
    response = ". ".join([s.strip() for s in sentences if s.strip()]) + "."

    # Fallback
    if response.strip() == ".":
        response = "I'm not completely sure about that. Could you rephrase the question?"

    print("Chatbot:", response)
```

## **Code Explanation**

### **conversation_history**
Stores previous conversation to maintain context.

### **Prompt Engineering**
Guides GPT-2 to behave like an assistant.

### **model.generate()**
Generates text based on input.

Parameters:
- temperature → controls randomness  
- top_k, top_p → control diversity

### **Decoding**
Converts generated tokens into readable text.

---

### **Loop**
Ensures continuous interaction until exit condition.
